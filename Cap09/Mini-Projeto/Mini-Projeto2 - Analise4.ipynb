```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

```python
# Calcule a média de preço por marca e por veículo
df2 = df.copy()

#Ok, I could just print this df2
df2 = df2.groupby(['brand', 'vehicleType'])['price'].mean()

# Unpack data
df2_dict = dict(df2)
means = list(df2_dict.values())
index = list(df2_dict.keys())
brands = list(map(lambda out: out[0], index))
types = list(map(lambda out: out[1], index))

new_df = pd.DataFrame(columns = ['avgPrice','brand','type'])
new_df['brand'] = brands
new_df['type'] = types
new_df['avgPrice'] = means
new_df.set_index( np.arange(len(means)))
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
# Crie um Heatmap com Preço médio de um veículo por marca, bem como tipo de veículo
import seaborn as sns
import numpy as np


# Reshape in 2D
new_df.fillna(0)
new_df2 = new_df.pivot(index = 'brand', columns = 'type', values = 'avgPrice').fillna(0)

# Plot data
fig = plt.figure(figsize = (14, 14))
sns.heatmap(new_df2, annot = True, cmap="YlGnBu", fmt = ".2f", robust = True, vmin = 0, vmax = 20000)
```

```python
# Salvando o plot
fig.savefig("plots/Analise4/heatmap-price-brand-vehicleType_.png")
```
