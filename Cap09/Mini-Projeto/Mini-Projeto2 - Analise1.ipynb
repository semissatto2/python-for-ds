## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
#clean_data_path = "dataset\autos.csv"
clean_data_path = "autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
df.head()
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro

fig = plt.figure(figsize = (14,8))
plt.xlabel('Ano de registro', fontsize = 15)
plt.ylabel('KDE', fontsize = 15)
plt.title('Distribuição de veículos com base no ano de registro', fontsize = 30)
sns.distplot(df['yearOfRegistration'], color = 'green')
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/vehicle-distribution_.png")
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
fig = plt.figure(figsize = (14,8))
plt.xlabel('Tipo de veículo', fontsize = 15)
plt.ylabel('Preço do veículo', fontsize = 15)
plt.title('Preço do veículo x Tipo de veículo', fontsize = 30)
vehiclePrice = df['price']
vehicleType = df['vehicleType']
sns.boxplot(x = vehicleType, y = vehiclePrice)
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/price-vehicleType-boxplot_.png")
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 
df2 = vehicleType.value_counts()
fig = plt.figure(figsize = (14,8))
sns.barplot(x = df2.index, y = df2.values)
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/count-vehicleType_.png")
```
