```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

```python
print(list(df))
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
df2 = df['brand'].value_counts()
fig = plt.figure(figsize = (14,8))
plt.xlabel('Número de carros', fontsize = 13, labelpad = 20)
plt.ylabel('Marca do carro', fontsize = 13, labelpad = 20)
plt.title('Número de veículos pertencentes a cada marca', fontsize = 18, pad = 30)
sns.barplot(y = df2.index, x = df2.values, palette = 'muted')
```

```python
# Salvando o plot
fig.savefig(("plots/Analise2/brand-vehicleCount_.png"))
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
import numpy as np
df3 = df.copy()
df3 = df.groupby(['vehicleType', 'gearbox'])['price'].mean()
df4 = df.groupby(['vehicleType', 'gearbox'])['price'].std()
fig = plt.figure(figsize = (14,8))
# Testando
#df3.plot.bar()

# Unpack the data
dict_group = dict(df3)
dict_group2 = dict(df4)
means = list(dict_group.values())
index = list(dict_group.keys())
std = list(dict_group2.values())

# Plot bars
fig , ax = plt.subplots(figsize = (14,8))
ind = np.arange(len(means)/3)
width = 0.2
# Mean
rects1 = ax.bar(ind - width, means[::3], width, color='orchid', label='unspecified')
rects2 = ax.bar(ind, means[1::3], width, color='orange', label='automatik')
rects3 = ax.bar(ind + width, means[2::3], width, color='lime', label='manuell')
# Std
rects4 = ax.bar(ind - width, np.array(std[::3])/5, width/5, color='black', label='standard deviation', bottom = means[::3] - np.array(std[::3])/10)
rects5 = ax.bar(ind, np.array(std[1::3])/5, width/5, color='black', bottom = means[1::3] -  np.array(std[1::3])/10)
rects6 = ax.bar(ind + width, np.array(std[2::3])/5, width/5, color='black', bottom = means[2::3] - np.array(std[2::3])/10)

ax.set_ylabel('Preço médio', fontsize = 15)
ax.set_title('Preço médio dos veículos por tipo de veículo e tipo de caixa de câmbio', fontsize = 25, pad = 20)
ax.set_xticks(ind)
tick_labels = list(map(lambda out: out[0], index[::3]))
ax.set_xticklabels(tick_labels, fontsize = 15)
ax.legend()
```

```python
# Salvando o plot
fig.savefig("plots/Analise2/vehicletype-gearbox-price_.png")
```
