```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
print (list(df))
```

```python
import numpy as np
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
df2 = df.copy()
df2 = df2.groupby(['fuelType', 'gearbox'])['price'].mean()
print (df2)

# Unpack data
df2_dict = dict(df2)

means = list(df2_dict.values())
index = list(df2_dict.keys())

unspecified_means = means[::3]
automatik_means = means[1::3]
manuell_means = means[2::3]

# Plot the bards
fig, ax = plt.subplots(figsize = (14,8))
ind = np.arange(len(means)/3)
width = 0.25
offset = 0.05
rect1 = ax.bar(ind - (width+offset/2), unspecified_means, width, color = 'goldenrod', label = 'Unspecified')
rect2 = ax.bar(ind, automatik_means, width, color = 'indigo', label = "Automatik")
rect3 = ax.bar(ind + (width+offset/2), manuell_means, width, color = 'tomato', label = "Manuell")

ax.set_ylabel('Preço médio', fontsize = 15)
ax.set_title('Preço médio dos veículos por tipo de combustível e tipo de caixa de câmbio', fontsize = 25, pad = 20)
ax.set_xticks(ind)
tick_labels = list(map(lambda out: out[0], index[::3]))
ax.set_xticklabels(tick_labels, fontsize = 15)
ax.legend()

```

```python
# Salvando o plot
fig.savefig("plots/Analise3/fueltype-vehicleType-price_.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
import numpy as np
df3 = df.copy()
df3 = df3.groupby(['vehicleType', 'gearbox'])['powerPS'].mean()
print (df3)

# Unpack data
df3_dict = dict(df3)
means = list(df3_dict.values())
index = list(df3_dict.keys())
unspecified_means = means[::3]
automatik_means = means[1::3]
manuell_means = means[2::3]

# Plot the bars
fig, ax = plt.subplots(figsize = (14,8))
ind = np.arange(len(means)/3)
width = 0.25
offset = 0.05
rect1 = ax.bar(ind - (width+offset/2), unspecified_means, width, color = 'goldenrod', label = 'Unspecified')
rect2 = ax.bar(ind, automatik_means, width, color = 'indigo', label = "Automatik")
rect3 = ax.bar(ind + (width+offset/2), manuell_means, width, color = 'tomato', label = "Manuell")

ax.set_ylabel('Potência média', fontsize = 15)
ax.set_title('Potência média dos veículos por tipo de combustível e tipo de caixa de câmbio', fontsize = 25, pad = 20)
ax.set_xticks(ind)
tick_labels = list(map(lambda out: out[0], index[::3]))
ax.set_xticklabels(tick_labels, fontsize = 15)
ax.legend()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/vehicletype-fueltype-power_.png")
```
